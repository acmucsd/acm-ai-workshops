###4. Loops

Loops help us repeat certain lines of code as many times as we need, so we don't have to write them multiple times

Say we wanted to find the mean of an array without Numpy, we would have to find the sum of all elements first then divide. Doing arr0[0]+arr0[1] + ... is way too inconvenient, and our code won't work for an array which different number of elements. That is why we need a loop.

````python
arr0 = [24, 48, 19, 392, 48, 78]
````

````python
i = 0 # index of the current array element, beginning at 0
sum = 0 # running total of the sum

# len() tells you how many elements are in the array
while i<len(arr0): # repeat code below while the given condition is true
  sum = sum + arr0[i]
  i+=1

# divide sum by number of elements to find mean
sum/len(arr0)
````

There are many ways to write a loop. Examples below achieve the same results, so it is up to you to choose a way that you like more

````python
sum = 0
for i in range(len(arr0)): # for each i in range from 0 up to len(arr0), repeat (i is incremented by 1 every time the loop runs)
  sum = sum + arr0[i]

sum/len(arr0)
````

````python
sum = 0
for elem in arr0: # for each element in the array, repeat (goes through all elements in the array in order)
  sum += elem

sum/len(arr0)
````

We can create nested loops when we traverse complex things like an array with multiple dimensions

````python
arr = [[25, 14, 23, 1, 5],
       [16, 17, 3, 19, 24],
       [4, 11, 25, 21, 18],
       [13, 2, 12, 9, 22]]
````

````python
sum = 0
for row in arr:
  for col in row:
    sum += col

sum/(len(arr)*len(arr[0]))
````

### 5. Functions

Functions are just reusable blocks of code. Built-in Python functions that we saw are print(), len() and input() - and there are many others. We can define our own functions as well

Previously we found a mean of 2-D array. Let's make it a function so that we don't have to retype this big chunk of code every time we need a mean

````python
myarr = [[25, 14, 23, 1, 5],
       [16, 17, 3, 19, 24],
       [4, 11, 25, 21, 18],
       [13, 2, 12, 9, 22]]
````

````python
def a2D_array_mean(arr):
  sum = 0
  for row in arr:
    for col in row:
      sum += col

  return sum/(len(arr)*len(arr[0])) # return is what function will give back to you
````

````python
# Let's try it on our myarr
print(a2D_array_mean(myarr))
````

## Intro to Data Preprocessing

Data preprocessing is how raw data is converted into a format that can be fed into a machine learning algorithm. It can involve steps like cleaning data, removing outliers, normalizing data, and splitting the data into training and test sets.

In the following section, we will download a dataset containing images of handwritten digits and preprocess it.

First, let's get the machine learning device set up. The following cell checks if a GPU is available and sets it as the device to speed up training; otherwise, it uses the CPU. It also sets up basic settings for loading training and testing data, using faster options if a GPU is available.

````python
import torch
from torchvision import datasets, transforms

# Check if a CUDA-enabled GPU is available
use_cuda = torch.cuda.is_available()

# Set batch sizes for training and testing
BATCH_SIZE = 50
TEST_BATCH_SIZE = 50

# Create dictionaries containing training and testing keyword arguments
train_kwargs = {"batch_size": BATCH_SIZE}
test_kwargs = {"batch_size": TEST_BATCH_SIZE}

# Set CUDA-specific keyword arguments if a GPU is available
if use_cuda:
    cuda_kwargs = {
        # Number of worker processes for loading data
        'num_workers': 1,
        # Pin data to GPU memory for faster transfer
        'pin_memory': True,
        # Shuffle data for each epoch during training
        'shuffle': True
    }
    # Set the device to use GPU
    device = torch.device("cuda")
    # Update training and testing keyword arguments with CUDA-specific arguments
    train_kwargs.update(cuda_kwargs)
    test_kwargs.update(cuda_kwargs)
else:
    # Set the device to use CPU
    device = torch.device("cpu")
````

The following cell uses transforms.Compose to chain together two transformations. The first, ToTensor, converts the image data into lists of numbers. The second, Normalize, scales the pixel values to a mean of 0.1307 and a standard deviation of 0.3081, which for the MNIST dataset will make the pixel values fall into a standard normal distribution.

````python
transform = transforms.Compose([
    transforms.ToTensor(),      # Convert image to PyTorch tensor
    transforms.Normalize((0.1307,), (0.3081,))  # Normalize pixel values
])
````

## Load the dataset and apply transformations



The following cell will download the train and test datasets with our transforms applied.

````python
train_dataset = datasets.MNIST(root='./data/MNIST', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data/MNIST', train=False, download=True, transform=transform)
````

With the dataset we have now, we want to specify how data will be loaded when training the ML model.

When training, we don't want to load the ENTIRE dataset all at once. That would be too much.

Instead, we want to use "mini-batches" such that only a "subset" of the dataset is loaded in each time. You might want to look into [the difference between batch vs epoch vs iteration](https://machinelearningmastery.com/difference-between-a-batch-and-an-epoch/#:~:text=The%20batch%20size%20is%20a%20number%20of%20samples%20processed%20before,samples%20in%20the%20training%20dataset.).

````python
# @title Solution
train_loader = torch.utils.data.DataLoader(train_dataset, BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, TEST_BATCH_SIZE, shuffle=False)
````
